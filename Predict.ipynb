```python
from keras import models
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt

drive_dir = "/content/drive"
```

```python
def predictions(image, labels):
  predict = model.predict(image)
  prediction_list.append(predict[0][0])

  index = np.argmax(predict)
  predicted_class = labels[index]

  predicted_labels.append(predicted_class)
```

```python
def plot_img(img, pred):

  plt.xlabel(f"Model predicted: {pred}")
  plt.imshow(img)
  plt.show()
```

```python
test_path = f"{drive_dir}/MyDrive/Fruits_Classification/fruit_dataset/test_set"

if test_path is None:
  print("Error to Read Path!")
else:
  print(test_path)

model = models.load_model(f"{drive_dir}/MyDrive/Fruits_Classification/fruitClassifier.h5")

true_labels=[]
labels = ["Apple", "Banana", "Strawberry"]
prediction_list=[]
predicted_labels=[]

for classes in sorted(os.listdir(test_path)):
  class_folders = os.path.join(test_path, classes)

  for img_files in sorted(os.listdir(class_folders)):
    img = cv2.imread(os.path.join(class_folders, img_files))

    true_labels.append(classes)

    if img is None:
      print("Error To Read Images!")

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    image = np.float64(cv2.resize(img,(64,64)))/255
    image = np.expand_dims(image, axis=0)

    predictions(image, labels)

    plot_img(img, predicted_labels)

cv2.destroyAllWindows()
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print("Confusion Matrix:\n",confusion_matrix(true_labels,predicted_labels))
print("Report\n",classification_report(true_labels,predicted_labels))
```

```python
def predict_img(path):

  lable = ["Apple", "Banana", "Strawberry"]

  img = cv2.imread(path)
  print(type(img))

  img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  image = np.float64(cv2.resize(img,(64,64)))/255.0
  image = np.expand_dims(image, axis=0)

  predict = model.predict(image)

  index = np.argmax(predict)
  pred = lable[index]

  plt.figure(figsize=(10,3))
  plt.xlabel(f"Model predicted: {pred}")
  plt.imshow(img)
  plt.show()
```
